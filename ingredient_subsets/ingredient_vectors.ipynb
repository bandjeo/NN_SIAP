#### Count how many ingredients failed to parse and purge them

```python
nulls = 0
not_nulls = 0
dataset_no_nulls = []
for recipe in dataset:
    no_nulls = True
    for ingr in recipe['ingredients_split']:
        if ingr == None:
            nulls += 1
            no_nulls = False
        else:
            not_nulls += 1
#     recipe['ingredients_split'] = [ingr for ingr in recipe['ingredients_split'] if ingr != None]
    if no_nulls:
        dataset_no_nulls.append(recipe)
# print(nulls)
# print(not_nulls)
# print(nulls/(nulls + not_nulls) * 100)
print(f'Length of whole dataset: {len(dataset)}')
print(f'Length of dataset with no nulls: {len(dataset_no_nulls)}')
```

```python
dataset = dataset_no_nulls
```

```python
import re
for recipe in dataset:
#     recipe['ingredients_combined'] = ' '.join([ingredient['ingredient'] for ingredient in recipe['ingredients_split']])
# this option combines recipe names by removing whitespaces
    recipe['ingredients_combined'] = []
    for ingredient in recipe['ingredients_split']:
        ingredient_one_word = ''.join(ingredient['ingredient'].lower().split())
        ingredient_one_word = re.sub("([\(\[]).*?([\)\]])|-|\d|[\(\)]", "", ingredient_one_word)
        ingredient_one_word = re.sub("-", "", ingredient_one_word)
        if ingredient_one_word:
            recipe['ingredients_combined'].append(ingredient_one_word)
```

```python
dataset[14]['ingredients_combined']
```

```python
from gensim.models import Word2Vec
ingredients_per_recipe = [recipe['ingredients_combined'] for recipe in dataset]
model = Word2Vec(sentences=ingredients_per_recipe, size=100, window=3, min_count=50, workers=4)
```

```python
print(model.wv.most_similar('pepper'))
print(model.wv['pepper'].shape)
```

```python
from scipy import spatial
first = "sugar"
second = "salt"
first_vec = model.wv[first]
second_vec = model.wv[second]
similarity = model.wv.cosine_similarities(first_vec, [second_vec])
print(similarity)
```

```python
import sys
word = "milk"
all_sims = model.most_similar(word, topn=sys.maxsize)
len_shows = 20
first_5 = list(all_sims[:len_shows])
last_5 = list(reversed(all_sims[-len_shows:]))
a = "\n\t".join([f'{x[0]:40} {x[1]}' for x in first_5])
b = "\n\t".join([f'{x[0]:39} {x[1]}' for x in last_5])
    
print(f'first 5: \n\t{a}\n\n')
print(f'last 5: \n\t{b}\n\n')
```
